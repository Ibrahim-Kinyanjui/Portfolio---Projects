```python
df['Last_Name'] = df['Last_Name'].str.strip('/')
df['Last_Name'] = df['Last_Name'].str.strip('...')
df['Last_Name'] = df['Last_Name'].str.strip('_')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('/', '-')
df['Phone_Number'] = df['Phone_Number'].str.replace('|', '-')
df
```

```python
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('N-a', 'NaN')
df
```

```python
df[['Street Address', 'City', 'Zipcode']] = df['Address'].str.split(', ', expand=True)
df
```

```python
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Y', 'Yes')
df['Paying Customer'] = df['Paying Customer'].str.replace('N', 'No')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Y', 'Yes')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('N', 'No')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No/a', 'NaN')
df
```

```python
df['Street Address'] = df['Street Address'].str.replace('N/a', 'NaN')
df
```

```python
df.fillna('')
df
```

```python
df = df.fillna('')
df
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('NaN', '')
df
```

```python
df['Street Address'] = df['Street Address'].str.replace('NaN', '')
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('NaN', '')
df
```

```python
df = df.rename(columns={'First_Name':'First Name'})
df
```

```python
df = df.rename(columns={'Last_Name':'Last Name'})
df = df.rename(columns={'Phone_Number':'Phone Number'})
df = df.rename(columns={'Do_Not_Contact':'Do Not Contact'})
df = df.rename(columns={'Not_Useful_Column':'Not Useful'})
df
```

```python
df.to_excel('Cleaned Customer Call List.xlsx', index=False)
```
